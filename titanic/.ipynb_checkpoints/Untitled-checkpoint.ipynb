```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import roc_auc_score

import pandas as pd
import matplotlib.pyplot as plt

X = pd.read_csv('train.csv')
y = X.pop("Survived")
```

```python
X.describe()
```

```python
X["Age"].fillna(X.Age.mean(), inplace=True)
```

```python
# Get only the numeric variables by selecting only the variables that are not of "object" datatypes
numeric_variables = list(X.dtypes[X.dtypes != "object"].index)
X[numeric_variables].head()
```

```python
def describe_categorical(X):
    from IPython.display import display, HTML
    display(HTML(X[X.columns[X.dtypes == "object"]].describe().to_html()))
```

```python
describe_categorical(X)
```

```python
# Drop the variables that won't be needed for this task
X.drop(["Name", "Ticket", "PassengerId"], axis=1, inplace=True)
```

```python
# Change the Cabin variable to be only the first letter or None
def clean_cabin(x):
    try:
        return x[0]
    except TypeError:
        return "None"

X["Cabin"] = X.Cabin.apply(clean_cabin)
```

```python
X.columns
```

```python
categorical_variables = ['Sex', 'Cabin', 'Embarked']

for variable in categorical_variables:
    # Fill missing data with the word "Missing"
    X[variable].fillna("Missing", inplace=True)
    
    # Create array of dummies
    dummies = pd.get_dummies(X[variable], prefix=variable)
    
    # Update X to include dummies and drop the main variable
    X = pd.concat([X, dummies], axis=1)
    X.drop([variable], axis=1, inplace=True)
```

```python
# Look at ll the columns in the dataset
def print_all(X, max_rows=10):
    from IPython.display import display, HTML
    display(HTML(X.to_html(max_rows=max_rows)))

print_all(X)
```

```python
model = RandomForestRegressor(100, oob_score=True, n_jobs=1, random_state=42)
model.fit(X, y)
print(roc_auc_score(y, model.oob_prediction_))
print(y.shape)
print(model.oob_prediction_.shape)
```

```python
model.feature_importances_
```

```python
# Simple version that shows all of the variables
feature_importances = pd.Series(model.feature_importances_, index=X.columns)
feature_importances.sort_values(inplace=True)
feature_importances.plot(kind="barh", figsize=(8, 6))
plt.show()
```
