```python
f = open('data.txt', 'r')
doc = f.read().split('\n')
print(doc)
```

```python
import random
```

```python
model = {}
for line in doc:
    # print(row['Quote'])
    line = line.split()
    for i, word in enumerate(line):
        if i == len(line)-1:   
            model['END'] = model.get('END', []) + [word]
        else:    
            if i == 0:
                model['START'] = model.get('START', []) + [word]
            model[word] = model.get(word, []) + [line[i+1]]

# len(model)
print(model['END'])
print(model['START'])
print(model)
```

```python
generated = []
while True:
    if not generated:
        words = model['START']
    elif generated[-1] in model['END']:
        if (len(generated) > 2):
            break
        else:
            generated = []
    else:
        words = model[generated[-1]]
    generated.append(random.choice(words))

print(generated)
```
